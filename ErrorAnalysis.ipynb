```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive
import math
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
file_path = 'Output/excludedVolume2.txt'
file = open(file_path)
lines = [line.rstrip('\n') for line in file]

T = list()
error = list()
```

```python
for line in lines:
    elements = line.split('\t')
    T.append(int(elements[0]))
    error.append(float(elements[1]))
```

```python
print('Analytical result 297')

plt.plot(T, error)
plt.axhline(0, color='black')
plt.xlabel('Timesteps')
plt.ylabel('Absolute Error')
```
